# flex

flex 默认值 0 1 auto

flex-grow 容器有剩余空间时是否允许弹性元素增大,以及相对其他同辈弹性元素的增大比例。其值必须是非负数，可以是小数。

`如果flex属性没有指定增长因子和缩减因子，则增长因子默认为1`。然而，如果flex和flex-grow都没指定,增长因子默认为0。

```python
%%html
<style>
  .container {
    display: flex;
    margin: 8px 0;
  }

  .container.flex1>div {
    flex-grow: 1;
    border: 1px solid red;
  }

  .container.flex2>div {
    flex: 1;
    border: 1px solid red;
  }
</style>
<div class="container flex1">
  <div>Test2</div>
  <div>Test22</div>
  <div>Test2222</div>
</div>
<div class="container flex2">
  <div>Test2</div>
  <div>Test22</div>
  <div>Test2222</div>
</div>
```
